Análise

Qual o tipo de alimento mais doado?

```python
contagem_alimentos = df_doacoes['alimento'].value_counts()

plt.figure(figsize=(10, 6))

contagem_alimentos.plot(kind='bar', color='skyblue', edgecolor='black')


plt.title('Contagem dos Alimentos Doações', fontsize=16)
plt.xlabel('Alimento', fontsize=12)
plt.ylabel('Contagem', fontsize=12)


plt.xticks(rotation=45, ha='right')


plt.tight_layout()
plt.show()
```

Qual a o tipo de comércio consegue realizar mais doações?

```python
agrupar = df_doacoes.groupby('comércio')['alimento'].apply(list).reset_index()
print(agrupar)
```

```python
agrupar['contagem_alimentos'] = agrupar['alimento'].apply(len)

plt.figure(figsize=(8, 5))
plt.bar(agrupar['comércio'], agrupar['contagem_alimentos'], color='green', edgecolor='black')


plt.title('Doações alimentos por comércio', fontsize=16)
plt.xlabel('Comércio', fontsize=12)
plt.ylabel('Número de Alimentos Doado', fontsize=12)


plt.tight_layout()
plt.show()
```

```python
contagem = df_doacoes.groupby('comércio')['alimento'].count().reset_index
print(contagem)
```

```python
agrupado_quantidade = df_doacoes.groupby(['comércio', 'alimento'])['quantidade'].sum().reset_index()
print(agrupado_quantidade)
```

```python
quantidade_total_alimentos = df_doacoes.groupby('alimento')['quantidade'].sum().reset_index
print(quantidade_total_alimentos)

```

Em quais períodos dos anos a doações são mais frequentes ?

```python
df_doacoes['mes_doacao'] = df_doacoes['data da doação'].dt.month
doacoes_por_mes = df_doacoes.groupby('mes_doacao')['quantidade'].sum().reset_index()
print(doacoes_por_mes)
```

```python

plt.figure(figsize=(8, 6))
plt.bar(doacoes_por_mes['mes_doacao'], doacoes_por_mes['quantidade'], color='blue', edgecolor='black')


plt.title('Total de Doações por Mês', fontsize=16)
plt.xlabel('Mês', fontsize=12)
plt.ylabel('Quantidade Total Doadas', fontsize=12)
plt.xticks(ticks=range(1, 13), labels=['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])


plt.show()
```

Análise de Outliers
Objetivo: Identificar doações incomuns, como quantidades muito altas ou datas de validade extremamente curtas ou longas.

Como fazer: Usar métodos estatísticos para detectar outliers em quantidade de doação ou dias até a validade.

```python
quantidade_outliers = df_doacoes[df_doacoes['quantidade'] > df_doacoes['quantidade'].mean() + 2 * df_doacoes['quantidade'].std()]
print(quantidade_outliers)

# Criar gráfico de barras para visualizar as quantidades
plt.figure(figsize=(10, 6))

# Gráfico de barras para todas as doações
df_doacoes['quantidade'].plot(kind='bar', color='lightgray', edgecolor='black', label='Quantidade Total')

# Destacar outliers com outra cor
plt.scatter(quantidade_outliers.index, quantidade_outliers['quantidade'], color='red', s=100, label='Outliers')

# Adicionar título e rótulos
plt.title('Visualização de Outliers em Quantidade de Doações', fontsize=16)
plt.xlabel('Índice', fontsize=12)
plt.ylabel('Quantidade', fontsize=12)

# Adicionar legenda
plt.legend()

# Ajustar layout para melhor visualização
plt.tight_layout()

# Exibir gráfico
plt.show()
```
